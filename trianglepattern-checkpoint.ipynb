```python
#print diagonal
n=5
for i in range(n):
    for j in range(n):
        if i==j or (i+j==n-1):
            print('*',end="  ")
        else:
            print(' ',end="  ")
    print()
```

```python
#square selectiom
n=5
for i in range(n):
    for j in range(n):
        if i>=j:
            print(i+1,end="  ")
        else:
            print(' ',end="  ")
    print()
```

```python
#Triangle selectiom
#lower triangle
n=5
for i in range(n):
    for j in range(n):
        if i>=j:
            print(i+1,end="  ")
        else:
            break
            print(' ',end="  ")
    print()
```

```python
#lower triangle
n=5
for i in range(n):
    for j in range(n):
        if i>=j:
            print(j+1,end="  ")
        else:
            print(' ',end="  ")
    print()
```

```python
#upper triangle
#hallow ( + print)
n=5
for i in range(n):
    for j in range(n):
        if i==j or j==0 or i==n-1:
            print('*',end="  ")
        else:
            if i>=j:
                print('+ ',end="  ")
    print()
```
